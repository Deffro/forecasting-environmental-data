```python
import matplotlib.pyplot as plt
import sys
sys.path.append('../../src/')
from functions import *
```

```python
stats = pd.DataFrame()

for dataset_name in ['Solcast', 'ERA5', 'ORAS5', 'AQPiraeus', 'Jena', 'Satellite', 'NOAA']:

    data, frequency_yearly_period, freq_sktime = read_file(dataset_name, data_path='H:/My Drive/PhD/ECOSCOPE/time-series-forecasting-waves/data/')

    # split data
    train, test, valid, train_without_valid, train_test_split_date, train_valid_split_date = train_valid_test_split(dataset_name, data)

    for target in data.columns:
        stats = stats.append({
            'Dataset': dataset_name,
            'Variable': target,
            'Train Size': train[target].shape[0],
            'Test Size': test[target].shape[0],
            'Mean': train[target].mean(),
            'Std': train[target].std(),
            'Min': train[target].min(),
            'Max': train[target].max(),
            'Has Trend': str(trend_test(train[target], variable_name=target)),
            'Has Seasonality': str(seasonality_test(train[target], variable_name=target, seasonal_period=frequency_yearly_period)),
            'Is Stationary': str(check_stationarity(train[target], variable_name=target))
        }, ignore_index=True)
        for c in ['Train Size', 'Test Size']:
            stats[c] = stats[c].astype(int)
```
